# Penghapusan anomaly data

```python
import pandas as pd
import folium
import folium.plugins
from folium.plugins import HeatMap
from folium.plugins import MarkerCluster
from IPython.display import display

# Import dataset
df = pd.read_csv('/content/hasil_preprocessing.csv')

# Batas geografis Kota Bandung (latitude dan longitude)
latitude_min, latitude_max = -6.973, -6.850
longitude_min, longitude_max = 107.560, 107.720

# Filter data untuk menentukan anomali
df_filtered = df[
    (df['Latitude'] >= latitude_min) & (df['Latitude'] <= latitude_max) &
    (df['Longitude'] >= longitude_min) & (df['Longitude'] <= longitude_max)
]

# Plot data asli
map_original = folium.Map(location=[-6.914744, 107.609810], zoom_start=12)
for idx, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Alamat Kejadian']).add_to(map_original)
# Plot data yang sudah difilter
map_filtered = folium.Map(location=[-6.914744, 107.609810], zoom_start=12)
for idx, row in df_filtered.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Alamat Kejadian']).add_to(map_filtered)

map_original.save('map_original.html')
display(map_original)

```

```python
map_filtered.save('map_filtered.html')
display(map_filtered)
```
